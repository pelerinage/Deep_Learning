Import Word2Vec model, i.e., collection of embedding vectors (vectors.bin file) into spaCy. two short sentences (5 to 6 words each). Compare their similarity using large English model provided by spaCy and imported Word2Vec model. Describe your observations.

```python
import spacy
```

```python
#load space english large model
nlp = spacy.load("en_core_web_lg")
```

```python
#define sentences to be compared
sen1 = "Every rational potency admits equally of contrary results, but irrational potencies admit of one result only."
```

```python
sen2 = 'But of beauty, I repeat again that we saw her there shining in company with the celestial forms; and coming to earth we find her here too, shining in clearness through the clearest aperture of sense.'
```

```python
#create docs for comparison
doc1 = nlp(sen1)
doc2 = nlp(sen2)
```

```python
#compare sentences with spacy engl model
similarity1 = doc1.similarity(doc2)
print(similarity1)
```

```python
#load word2vec model that was generated by:  spacy init vectors en vectors.bin /Users/Matt/CS89
nlp2 = spacy.load('/Users/Matt/CS89')
```

```python
doc1b = nlp2(sen1)
doc2b = nlp2(sen2)
```

```python
#compare sentences with spacy engl model
similarity2 = doc1b.similarity(doc2b)
print(similarity2)
```

We see a very drastic difference in the similarity values between the two models.  The spacy large english model rated the sentences from aristotle and plato as nearly 85% similar.  However, the word2vec vectors.bin model only identified the sentences as 42% similar.
